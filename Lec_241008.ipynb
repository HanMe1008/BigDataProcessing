```python
### 빅데이터처리:해양환경 
## 해양학전공 전상규 202013297

### 실습 6
import numpy as np
import matplotlib.pyplot as plt

spread = np.random.rand(50) * 100
center = np.ones(25) * 50
filter_high = np.random.rand(10) * 100 + 100
filter_low = np.random.rand(10) * -100

data = np.concatenate((spread, center, filter_high, filter_low))

plt.figure(figsize=(8,8))
plt.boxplot(data)

plt.xlabel('Sample', fontsize=13)
plt.title('Box Plot', fontsize=13)

plt.gca().set_facecolor('white')

plt.show()
```

```python
### 빅데이터처리:해양환경 
## 해양학전공 전상규 202013297

### 실습 7
import numpy as np
import matplotlib.pyplot as plt

data1 = np.concatenate((np.random.rand(50) * 100,
                        np.ones(25) * 50,
                        np.random.rand(10) * 100 + 100,
                        np.random.rand(10) * -100))

data2 = np.concatenate((np.random.rand(50) * 100,
                        np.ones(25) * 50,
                        np.random.rand(10) * 100 + 100,
                        np.random.rand(10) * -100))

# 박스플롯의 설정
num_group = 2
edgecolor = 'blue'                # 박스 경계 색상
linewidth = 2                     # 박스 경계 선 두께
facecolors = ['#32a852', '#8e59d9']  # 박스 내부 색상
medianlinecolor = 'white'         # 중앙값 선 색상

fig=plt.figure(figsize=(8,8))
fig.set_facecolor('white')

box = plt.boxplot([data1, data2], patch_artist=True, boxprops=dict(linewidth=2, edgecolor='blue'))
 
for item in ['boxes', 'fliers', 'medians']:
    for i, v in enumerate(box[item]):   # 각 요소의 인덱스(i)와 값(v)을 가져옴
        if item == 'boxes':
            plt.setp(v, facecolor=facecolors[i], edgecolor=edgecolor)
        elif item == 'fliers':
            plt.setp(v, markerfacecolor=facecolors[i])
        else:
            plt.setp(v, color=medianlinecolor)

for item in ['whiskers', 'caps']:
    for i, v in enumerate(zip(box[item][::2],box[item][1::2])):
        if item == 'whiskers':
            plt.setp(v, color=edgecolor, linewidth=linewidth, linestyle='--')
        else:
            plt.setp(v, color=edgecolor, linewidth=linewidth)

plt.title('Box Plot', fontsize=13)
plt.xticks([1, 2], ['Group 1', 'Group 2'], fontsize=13)

plt.show()
```
