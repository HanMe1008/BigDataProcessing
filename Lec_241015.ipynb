```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

### 1. 데이터 준비

data = {'Location': ['St.1', 'St.1', 'St.1', 'St.2', 'St.2', 'St.2', 'St.3', 'St.3', 'St.3'],
        'Temperature': [22.5, 23.1, 22.8, 24.2, 24.5, 24.1, 20.0, 20.5, 19.8],
        'Salinity': [35.1, 35.3, 35.0, 34.8, 34.9, 35.2, 33.5, 33.8, 33.7],
        'Oxygen': [6.2, 6.5, 6.3, 5.9, 6.1, 6.0, 7.1, 7.0, 7.2]}

df = pd.DataFrame(data)

print(df)
```

```python
### 실습 8

plt.figure(figsize=(12,8))

# 수온
plt.subplot(1, 3, 1)
sns.boxplot(x='Location', y='Temperature', data=df)
plt.title('Temperature by Location')
plt.ylabel('Temperature (°C)')

# 염분
plt.subplot(1, 3, 2)
sns.boxplot(x='Location', y='Salinity', data=df)
plt.title('Salinity by Location')
plt.ylabel('Salinity (psu)')

# 용존산소
plt.subplot(1, 3, 3)
sns.boxplot(x='Location', y='Oxygen', data=df)
plt.title('Oxygen by Location')
plt.ylabel('Oxygen (mg/L)')

plt.tight_layout()
plt.show()

```

```python
### 실습 9 
## 색상 팔레트(color palette) 사용
palette = sns.color_palette("Set2")

plt.figure(figsize=(12,8))

# 수온
plt.subplot(1, 3, 1)
sns.boxplot(x='Location', y='Temperature', data=df, hue='Location', palette="Set2", dodge=False)
plt.title('Temperature by Location')
plt.ylabel('Temperature (°C)')

# 염분
plt.subplot(1, 3, 2)
sns.boxplot(x='Location', y='Salinity', data=df, hue='Location', palette="Set2", dodge=False)
plt.title('Salinity by Location')
plt.ylabel('Salinity (psu)')

# 용존산소
plt.subplot(1, 3, 3)
sns.boxplot(x='Location', y='Oxygen', data=df, hue='Location', palette="Set2", dodge=False)
plt.title('Oxygen by Location')
plt.ylabel('Oxygen (mg/L)')

plt.tight_layout()
plt.show()

```

```python
### 실습 10
## 색상 팔레트(color palette) 사용 + 주석(annotation) 추가

palette = sns.color_palette("Set2")

plt.figure(figsize=(15,8))

# 수온
plt.subplot(1, 3, 1)
sns.boxplot(x='Location', y='Temperature', data=df, hue='Location', palette="Set2", dodge=False)
plt.title('Temperature by Location')
plt.ylabel('Temperature (°C)')

# 수온 주석
plt.text(0, df['Temperature'].mean()+1,
         'St.1 has the highest temperature range \nand variability.',
         ha='center', va='bottom')
plt.text(1, df['Temperature'].mean()+1.65,
         'St.2 has moderate temperature values. \nLess variability compared to St.1.',
         ha='center', va='bottom')
plt.text(2, df['Temperature'].mean()-1.9,
         'St.3 has the lowest temperatures. \nLess variability in this location.',
         ha='center', va='bottom')

# 염분
plt.subplot(1, 3, 2)
sns.boxplot(x='Location', y='Salinity', data=df, hue='Location', palette="Set2", dodge=False)
plt.title('Salinity by Location')
plt.ylabel('Salinity (psu)')

# 염분 주석
plt.text(0, df['Salinity'].mean()+0.05,
         'St.1 has a slightly higher salinity range \ncompared to others.',
         ha='center', va='bottom')
plt.text(2, df['Salinity'].mean()+0.7,
         'St.2 has a narrower salinity range. \nModerate values overall.',
         ha='center', va='bottom')
plt.text(2, df['Salinity'].mean()-0.7,
         'St.3 has the lowest salinity values. \nThe range is fairly narrow.',
         ha='center', va='bottom')

# 용존산소
plt.subplot(1, 3, 3)
sns.boxplot(x='Location', y='Oxygen', data=df, hue='Location', palette="Set2", dodge=False)
plt.title('Oxygen by Location')
plt.ylabel('Oxygen (mg/L)')

# 용존산소 주석
plt.text(0, df['Oxygen'].mean()+0.25,
         'St.1 has a slightly higher salinity range \ncompared to others.',
         ha='center', va='bottom')
plt.text(1, df['Oxygen'].mean()-0.25,
         'St.1 has a slightly higher salinity range \ncompared to others.',
         ha='center', va='bottom')
plt.text(2, df['Oxygen'].mean()+0.65,
         'St.1 has a slightly higher salinity range \ncompared to others.',
         ha='center', va='bottom')

plt.tight_layout()
# plt.subplots_adjust(wspace=0.5, hspace=0.5) 
plt.show()

```
