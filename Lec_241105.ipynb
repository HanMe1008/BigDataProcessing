```python
## Basemap 사용 !

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
from netCDF4 import Dataset
import netCDF4 as nc 

data = nc.Dataset('F:\BigDataProcessing\KHOA_SST_20231008.nc')
print(data['sst'],"\n")

lons = data['lon'][:].data
lats = data['lat'][:].data
sst  = data['sst'][:].data

# Basemap 객체 생성
m = Basemap(projection='merc', llcrnrlat=np.min(lats), urcrnrlat=np.max(lats),
            llcrnrlon=np.min(lons), urcrnrlon=np.max(lons), resolution='i')
# ur : upper right,  ll : lower left

plt.figure(figsize=(10,10))

# 지도 표시
m.drawcoastlines()
m.drawcountries()
# 위도와 경도 표시(2도 간격)
m.drawparallels(np.arange(np.floor(np.min(lats)), np.ceil(np.max(lats)), 2), labels=[1, 0, 0, 0])   # [남쪽, 중앙, 북쪽, 극북쪽]
m.drawmeridians(np.arange(np.floor(np.min(lons)), np.ceil(np.max(lons)), 2), labels=[0, 0, 0, 1])   # [서쪽, 중앙, 동쪽, 극동쪽]

# 경도 및 위도 그리드
lon, lat = np.meshgrid(lons, lats)
x, y = m(lon, lat)

# SST 데이터를 이미지로 표시
m.pcolormesh(x, y, np.squeeze(sst), cmap='jet')

plt.colorbar(label='Sea Surface Temperature (°C)')
plt.title('Sea Surface Temperature in the NorthWest Pacific Ocean')

plt.show()
```

```python
## Basemap 사용 (2)

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
from netCDF4 import Dataset
import netCDF4 as nc 

# Load the NetCDF data
data = nc.Dataset('F:\BigDataProcessing\KHOA_SST_20231008.nc')

# Extract Longitude, Latitude, and SST data
lons = data['lon'][:]
lats = data['lat'][:]
sst  = data['sst'][:]

# Handle dimensions if necessary
# if sst.ndim == 3:
#     sst = sst[0]

# Create a Basemap instance
m = Basemap(projection='merc',
            llcrnrlat=np.min(lats),
            urcrnrlat=np.max(lats),
            llcrnrlon=np.min(lons),
            urcrnrlon=np.max(lons),
            resolution='i')
"""
llcrnrlat : lower left corner latitude
ur : upper right 
llcrnrlon : lower left corner longitude
ll : lower left
"""

plt.figure(figsize=(10,10))

# Draw coastlines and countries 
m.drawcoastlines()
m.drawcountries()

# Adjust the spacing of parallels and meridians
parallels_interval = 3  # Latitude interval
meridians_interval = 3  # Longitude interval

# Generate latitude and longitude values
parallels = np.arange(np.floor(np.min(lats)), np.ceil(np.max(lats)), parallels_interval)
meridians = np.arange(np.floor(np.min(lons)), np.ceil(np.max(lons)), meridians_interval)

# Draw parallels
m.drawparallels(parallels, labels=[1, 0, 0, 0], fontsize=6)

# Draw meridians
m.drawmeridians(meridians, labels=[0, 0, 0, 1], fontsize=6)  

# Create grid for mapping
lon, lat = np.meshgrid(lons, lats)
x, y = m(lon, lat)

# Plot SST data
c = m.pcolormesh(x, y, np.squeeze(sst), cmap='jet', shading='auto')

# Add colorbar and title
plt.colorbar(c, label='Sea Surface Temperature (°C)')
plt.title('Sea Surface Temperature in the NWPO (08.Oct.2023)', fontsize=12)

# Show the plot
plt.show()
```

```python
## NetCDF 저장

# 1. 파일 경로 정의
output_file = 'F:\BigDataProcessing\KHOA_SST_20231008_processed.nc'

# 2. NetCDF 파일 생성
with Dataset(output_file, 'w', format='NETCDF4') as new_data:
    # 3. 차원 정의
    new_data.createDimension('lat', len(lats))
    new_data.createDimension('lon', len(lons))

    # 4. 변수 생성
    latitudes = new_data.createVariable('lat', 'f4', ('lat',))
    longitudes = new_data.createVariable('lon', 'f4', ('lon',))
    sea_surface_temp = new_data.createVariable('sst', 'f4', ('lat', 'lon',))

    # 5. 변수에 데이터 할당
    latitudes[:] = lats
    longitudes[:] = lons
    sea_surface_temp[:] = sst

    # 6. 속성 추가
    latitudes.units = 'degrees_north'
    longitudes.units = 'degrees_east'
    sea_surface_temp.units = 'degrees_Celsius'

    # 7. 글로벌 속성 추가
    new_data.title = 'Processed Sea Surface Temperature Data'
    new_data.institution = 'Your Institution'
    new_data.source = 'Original data source information' 
    new_data.history = 'Created' + np.datetime_as_string(np.datetime64('now'), unit='s')

# 8. 결과 출력
print(f"Data saved to {output_file}")
```
